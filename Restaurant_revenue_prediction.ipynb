### Remove the date column

```python
df=df.drop(columns=['Open Date'])
```

### Convert Categorical Features to Numerical Features by creating dummies

```python
new_data=pd.get_dummies(df,drop_first=True)
column_list=list(new_data.columns)
```

```python
new_data
```

### Seperate the Predictor from the label

```python
# feature data (predictors)
X = new_data.iloc[:, :-1]
# label class
y = df['revenue']
```

### Standardize Data

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

### Partition dataset into train and test set

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.30, random_state=42)
```

```python
# Model training
model = XGBRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
rmse = sqrt(mean_squared_error(y_test, y_pred))
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
